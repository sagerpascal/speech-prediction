```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

# Read CSV

```python
base_path = Path('dfs')
files = ["timit-orig-train.csv", "timit-orig-val.csv", "timit-orig-test.csv"]
dataframes = []

for f in files:
    dataframes.append(pd.read_csv(base_path / f))

dataframes[2].head()
```

```python
for i, d in enumerate(dataframes):
    h = d.plot(kind='hist', grid=True, bins=100, title="MFCC Frame Length", figsize=(15, 10))
    plt.tight_layout()
    #plt.show()
    plt.savefig("{}.png".format(files[i]))
```
